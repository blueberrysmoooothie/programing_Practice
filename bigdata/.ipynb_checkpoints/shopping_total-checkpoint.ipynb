### 네이버 자료 대분류별 자료 통합 분석


    - 자료출처 : 네이버 데이터 랩 (https://datalab.naver.com/)
    - 조원별로 각 분야의 성별, 나이대별 자료를 통합하고 일정한 형식을 가진 데이터 프레임으로 변환

```python
TOTAL_PATH = './naver_shopping/'
os.listdir(TOTAL_PATH)
```

```python
total_datas = [pd.read_csv(TOTAL_PATH + f, index_col= 0 ) for f in os.listdir(TOTAL_PATH)]
total_data_names = [f.split('.')[0] for f in os.listdir(TOTAL_PATH)]
```

```python
total_datas[0]
```

```python
for data in total_datas:
    print(data.columns)
```

```python
for i in range(len(total_datas)):
    print(total_data_names[i],' :\n',total_datas[i].index[[0,-1]], '\n')
```

```python
total_total = pd.concat([data.apply(sum, axis = 1) for data in total_datas], axis = 1)
total_total.columns = total_data_names

total_total
```

```python
fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_total.iloc[:-1].plot(ax = ax)
```

```python

total_total.index = pd.to_datetime(total_total.index)
total_month_datas = total_total.to_period('m').reset_index().groupby('날짜').sum()


fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_month_datas.loc['2021-08': '2022-06'].plot(ax = ax)

plt.title('월 단위 분야별 클릭 횟수')
plt.xlabel('')
```

- 화장품이 가장 낮은 클릭수를 보인다.
- 도서분야가 변동이 크다고 판단
- 도서 분야에서 특정 월만 높은 관심도를 보이는데 이를 자세히 알아보고 해결 방안을 찾아보기

```python
def min_max(data):
    data_max, data_min = data.max(), data.min() 
    return (data-data_min)/(data_max - data_min)
    
total_month_datas.apply(min_max)
```

```python
fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_month_datas.loc['2021-08': '2022-06'].apply(min_max).plot(ax = ax)

plt.title('월 단위 분야별 클릭 횟수 - 정규화')
plt.xlabel('')
```

### 도서 분야별 네이버 자료 비교

- 도서분야에 대해 2차적인 분석 실시

        - 도서분야 자료내의 중분류들에 대해서도 대분류에서의 과정과 마찬가지로 
        - 조원들이 나누어서 자료를 만들고 통합하여 전체 자료를 생성


```python
TOTAL_BOOK_PATH = './book_naver/'
os.listdir(TOTAL_PATH)


total_books = [pd.read_csv(TOTAL_BOOK_PATH + f, index_col= 0 ) for f in os.listdir(TOTAL_BOOK_PATH)]
total_book_names = [f.split('.')[0] for f in os.listdir(TOTAL_BOOK_PATH)]


total_total_books = pd.concat([data.apply(sum, axis = 1) for data in total_books], axis = 1)
total_total_books.columns = total_book_names

total_total_books
```

```python
for i in range(len(total_books)):
    print(total_book_names[i],' :\n',total_books[i].index[[0,-1]], '\n')
```

```python
fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_total_books.plot(ax = ax)

plt.title('도서 분야별 클릭 횟수')
plt.xlabel('')
```

```python

total_total_books.index = pd.to_datetime(total_total_books.index)
total_month_books = total_total_books.to_period('m').reset_index().groupby('날짜').sum()


fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_month_books.loc['2021-08': '2022-06'].plot(ax = ax)

plt.title('월 단위 도서 분야별 클릭 횟수')
plt.xlabel('')
```

대부분 12월, 3월에 많은 증가량을 보임.  
이는 년이 바뀌는 시기와 개학, 개강시즌에 의한 영향으로 판단 하였습니다.

```python
fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)
total_month_books.loc['2021-08': '2022-06'].apply(min_max).plot(ax = ax)
plt.title('월 단위 도서 분야별 클릭 횟수 - 정규화')
plt.xlabel('')
```

##### 출판 유통 통합 유통망 내 전체 자료 분석
    - 실제 판매 기록을 통해 네이버 쇼핑에서의 클릭수가 실제 고객의 관심사를 반영하는지 확인
    - 자료출처 : https://bnk.kpipa.or.kr/home/v3/addition/adiStatsCtgSaleCnt


```python
total_report = pd.read_csv('./book/총합_판매_보고서.csv', header = 2, index_col=0)
total_report.columns = total_report.columns.str.replace('.1', '_온라인')
total_report.columns = total_report.columns.str.replace('.2', '_지역서점')
total_report.rename(columns = {'부수' : '부수_오프라인', '종수' : '종수_오프라인','금액' : '금액_오프라인'}, inplace = True)
total_report
```

```python
total_report.replace(',', '', regex=True, inplace = True)
total_report = total_report.astype(float)
total_report.dtypes
```

```python

fig = plt.figure(figsize=(18, 10))
ax = plt.subplot(3,1,1)
total_report.filter(like = '부수').plot(ax = ax)
ax = plt.subplot(3,1,2)
total_report.filter(like = '종수').plot(ax = ax)
ax = plt.subplot(3,1,3)
total_report.filter(like = '금액').plot(ax = ax)
```

9월, 12월, 3월에  판매 부수와 판매 금액이 증가

### 네이버 쇼핑 클릭 수와 실재 판매 판수 간의 관계 분석

```python
for col in ['부수','금액']:
    total_report[col + '_total'] = total_report.filter(like = col).apply(sum, axis = 1)
    
total_report_total = total_report[total_report.columns[-2:]].copy()[1:]
total_report_total
```

```python
total_month_books_total = total_month_books.apply(sum, axis = 1)[1:-1]
total_month_books_total = total_month_books_total.to_frame()
total_month_books_total.columns = ['전체_클릭수']
total_month_books_total.index = total_report_total.index
total_month_books_total
```

```python
corr_df = pd.concat([total_report_total, total_month_books_total],axis = 1)
corr_df
```

```python
corr_df.corr()
```

0.69, 0.71로 높은 상관계수를 보인다.

```python
fig = plt.figure(figsize=(18, 8))
ax = plt.subplot(1,1,1)

corr_df.apply(min_max).plot(ax = ax)
plt.title('판매량, 금액, 클릭 수 정규화')

plt.xlabel('')
```

```python
import seaborn as sns
sns.pairplot(corr_df)
```

결론 
- 도서 분야에서 전체 클릭 수가 높은 경우는 판매 지표도 높은 결과를 보였다.
- 하지만 클릭 수가 낮은 경우에도 높은 판매 지표를 보이기도 한다.
- 이는 클릭 수가 높으면 판매 량도 높지만, 반드시 클릭 수가 많아야 판매량이 높은것이 아님을 의미한다.  
- 즉, 네이버 쇼핑에서의 클릭 수가 도서 구매 수에 대해 필요조건이 아닌 충분조건이라는 의미이다.
