```python
import pandas as pd
```

```python
hollys_branches = pd.read_csv('hollys_branches.csv', encoding= 'utf-8')
hollys_branches
```

```python
for i, x in hollys_branches.iterrows():
    print('''[{0:3}]: 매장이름: {1}, 지역: {2}, 주소: {3}, 전화번호: {4}'''.format(i+1, x['매장이름'], x['위치(시,구)'], x['주소'], x['전화번호']))
print('전체 매장 수: {}'.format(hollys_branches.shape[0]))
```

```python
# 검색어를 공백 단위로 분할
search_str = input('검색할 매장의 도시를 입력하세요:').split()

# 분할한 검색어를 모두 포함하는지 판단하는 조건생성
cond= [True] * (hollys_branches.shape[0])
for s in (search_str):
    cond *= hollys_branches['주소'].str.contains(s)
    
# 조건으로 걸러낸 데이터 검색 결과 프래임 생성
searched_center = hollys_branches[cond].copy().reset_index(drop = True)

print('--------------------')
print(f'검색된 매장 수: {searched_center.shape[0]}')
print('--------------------')
for i, x in searched_center.iterrows():
    print("[{0:3}]: ['{1}', '{2}']".format(i+1, x['주소'], x['전화번호']))
```
