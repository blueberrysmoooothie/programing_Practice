## 자료 전처리및 통합 과정

```python
import pandas as pd
LFproducts_csv = './농축수산물.csv'
covid19_xlsx = './코로나바이러스감염증-19_확진환자_발생현황_220718.xlsx'
oilPrice_csv = './원유가격현황.csv'
interestRate_csv = './국제 주요국 중앙은행 정책금리_19141249.csv'
```

#### 농축수산물 자료

```python

LFproducts = pd.read_csv(LFproducts_csv,index_col = 1, encoding = 'euc-kr').drop('시도별',axis = 1).T
LFproducts
```

```python
LFproducts.info()
```

```python
LFproducts.drop('(농산물)기타농산물',axis = 1, inplace = True)
LFproducts = LFproducts.rename(columns = {'(농산물)곡물':'곡물', '(농산물)채소':'채소', '(농산물)과실':'과실'})
```

```python
LFproducts.index
```

```python
LFproducts.index = pd.period_range('2019-01-01','2022-06-30',freq = 'm')
```

#### 코로나 환자 현황 자료

```python
covid19 = pd.read_excel('./코로나바이러스감염증-19_확진환자_발생현황_220718.xlsx', skiprows=4).drop(0)

covid19 = covid19.iloc[:,[0,1,4]].rename(columns = {'계(명)': '확진자 수', '사망(명)': '사망자 수'})
covid19
```

```python
covid19.info()
```

```python
covid19['확진자 수'].unique()
```

```python
covid19['사망자 수'].unique()
```

```python
covid19['사망자 수'] = covid19['사망자 수'].replace('-', 0)
```

```python
covid19['사망자 수'].unique()
```

```python
covid19.info()
```

```python
covid19['일자'] = pd.to_datetime(covid19['일자'])
covid19.set_index('일자', inplace = True)
covid19 = covid19.to_period('m')
covid19
```

```python
covid19 = covid19.reset_index().groupby('일자').sum()
```

```python
covid19 = covid19.drop('2022-07')
```

#### 기름 가격 변동 자료

```python
oilPrice = pd.read_csv(oilPrice_csv)
oilPrice = oilPrice[['날짜', '종가']]
oilPrice
```

```python
oilPrice.info()
```

```python
oilPrice['날짜'] = pd.to_datetime(oilPrice['날짜'], format = '%Y년 %m월 %d일')
oilPrice = oilPrice.set_index('날짜').to_period('m').reset_index().groupby('날짜').mean()
```

```python
oilPrice.columns = ['원유 종가']
oilPrice
```

#### 금리변동 자료

```python
interestRate = pd.read_csv(interestRate_csv, encoding = 'euc-kr', index_col = 1).iloc[:,3:].T
interestRate
```

```python
ys = ['2019']*12 +['2020']*12 +['2021']*12 + ['2022']*6
interestRate.index = [ys[i] +' '+ interestRate.index[i] for i in range(len(ys))]
interestRate.index = pd.to_datetime(interestRate.index, format = '%Y %b-%d')
interestRate = interestRate.to_period('m')[['한국']]
```

```python
interestRate.columns = ['한국 금리']
interestRate = interestRate.fillna(2.25)
interestRate
```

#### 데이터 통합

```python
LF_total = pd.concat([LFproducts, covid19, oilPrice, interestRate], axis = 1, ).fillna(0)
```

```python
LF_total
```

## 통합자료 정규화

```python
def get_range(s):
    return [s.max(), s.min(), s.max() - s.min()]
```

```python
LF_total.apply(get_range)
```

```python
# 코로나 자료 로그 변환
import numpy as np
scaled_LF = LF_total.copy()
scaled_LF[['확진자 수', '사망자 수']] = scaled_LF[['확진자 수', '사망자 수']].applymap(lambda x : np.log10(x+1))
scaled_LF.rename(columns = {'확진자 수':'log(확진자 수)', '사망자 수': 'log(사망자 수)'}, inplace = True)
scaled_LF
```

```python
scaled_LF.apply(get_range)
```

```python
def min_max_scal(x):
    return((x-x.min())/(x.max()-x.min()))
```

```python
scaled_LF = scaled_LF.apply(min_max_scal)
scaled_LF
```

## 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
import platform
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
    
plt.style.use('ggplot')
```

```python
lf_cols = list(scaled_LF.columns[:6])
fac_cols = list(scaled_LF.columns[6:])
for i in range(6):
    cols = [lf_cols[i]]+ fac_cols
    sns.pairplot(scaled_LF[list([lf_cols[i]]+ fac_cols)])
    plt.show()
```

농축수산물 전체와 각 분류별 소비자 물가의 변동을 시각화
- 채소, 과실, 축산물의 변동 폭이 크다.  

```python
colors = sns.color_palette('hls', 5)

def draw_scaled(scale_df, col):
    plt.figure(figsize=(16,6))
    plt.bar(range(len(scale_df.index)), scale_df[col], alpha = 0.5, color = colors[0])
    plt.xticks(range(len(scale_df.index)), scale_df.index, rotation=80)
    for fac in fac_cols:
        plt.plot(range(len(scale_df.index)),scale_df[fac],
                 'o--', color = colors[fac_cols.index(fac)+1])
    legs = list(scale_df.columns[6:]) +[col] 

    plt.legend(legs,loc = 2)
    
    plt.axvline(x=37, ymin=0, ymax=1.0, color = 'black', alpha = 0.5)
    plt.text(x=37, y = 0.93, s = '우-러 전쟁 발발 ', fontsize = 12, color = 'r', ha = 'right' )
    plt.title('정규화 된 '+ col +' 소비자 물가 변동 추이', fontsize = 18 )
    plt.show()

for col in lf_cols:
    draw_scaled(scaled_LF, col)
   
```

농축수산물 전체와 각 분류별 소비자 물가지수를 정규화 하고  
코로나 관련 수치들, 원유, 금리 수치들도 정규화 하여 전체 자료의 정규화된 변동을 시각화

곡물 분야를 제외하고 전체적으로 코로나 수치와 원유 종가에 따라 증가하는 경향을 보였다.

농축수산물 전체
- 코로나 발생 이후 지속적으로 증가하였다. 전쟁 발발 직후 감소하는듯 하였으나 증가 

곡물
- 변동폭이 적어 의미는 부족하나 2020년 말 급증한 후 다시 감소

채소
- 변동이 컸으며 이 그래프에서도 주기성이 보인다. 코로나 이후 전체적으로 증가한듯 하다.

과실
- 변동이 컸으며 코로나 이후 많은 증가를 보였다.

축산물
- 코로나환자, 사망자의 수치와 유사한 증가 경향을 보이며 코로나 이후 원유 종가와도 연관을 보인다.

수산물
- 실제 변동폭이 적어 정규화된 자료의 의미가 부족하다. 축산물과 유사한 경향을 보인다.

### 코로나 확진자 수와 농축수산물 소비자 물가 사이의 관계 상세분석

```python
target_data = LF_total[['농축수산물','확진자 수']]
target_data
```

```python
target_data['확진자 수'] = np.log10(target_data['확진자 수']+1)
target_data.rename(columns = {'확진자 수': 'log(확진자 수)'}, inplace = True)
target_data
```

```python
target_data['log(확진자 수)'] = target_data['log(확진자 수)']*3 + 90
target_data.rename(columns = {'log(확진자 수)': 'log(확진자 수)*3 + 90'}, inplace = True)
target_data
```

```python
target_data.plot()
```

코로나 확진자 수와 농축수산물의 소비자물가가 함께 증가하는 경향을 보인다.

```python
sns.pairplot(target_data)
plt.show()
sns.heatmap(target_data.corr(),annot= True,cmap = 'Blues')
plt.show()
```
